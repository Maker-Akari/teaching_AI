# sklearn 实现 neural network

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# 训练困难的 moon 数据

```python
from sklearn import datasets
X, y = datasets.make_moons(500, noise=0.15, random_state=1)

plt.scatter(X[:,0], X[:,1], c=y, cmap='rainbow')
```

```python
from sklearn import neural_network, model_selection, metrics
model = neural_network.MLPClassifier((50,20,),  )
model.fit(X,y)
y_predict = model.predict(X)
```

```python
model.get_params()
```

```python
fig, axs = plt.subplots(1,2, figsize=(14,6))
accuracy = metrics.accuracy_score(y, y_predict)
axs[0].scatter(X[:,0], X[:,1], c=y, cmap='rainbow'); axs[0].set_title('original data')
axs[1].scatter(X[:,0], X[:,1], c=y_predict, cmap='rainbow'); axs[1].set_title('ANN prediction (accuracy = %.3f'%accuracy+')')
axs[0].grid(); axs[1].grid()
```

```python
model_selection.cross_val_score(model, X, y, cv=5)
```

##### 如果 activation 取作 identity (没有激活)，就无法拟合非线性数据

```python
from sklearn import neural_network, model_selection, metrics
model = neural_network.MLPClassifier((50,20,), activation='identity' )
model.fit(X,y)
y_predict = model.predict(X)

fig, axs = plt.subplots(1,2, figsize=(14,6))
accuracy = metrics.accuracy_score(y, y_predict)
axs[0].scatter(X[:,0], X[:,1], c=y, cmap='rainbow'); axs[0].set_title('original data')
axs[1].scatter(X[:,0], X[:,1], c=y_predict, cmap='rainbow'); axs[1].set_title('ANN prediction (accuracy = %.3f'%accuracy+')')
axs[0].grid(); axs[1].grid()
```

# 训练 circles 数据

```python
X, y = datasets.make_circles(500, noise=0.1, factor=0.5, random_state=2)
model = neural_network.MLPClassifier((50,20,),)
model.fit(X,y)
y_predict = model.predict(X)

fig, axs = plt.subplots(1,2, figsize=(14,6))
accuracy = metrics.accuracy_score(y, y_predict)
axs[0].scatter(X[:,0], X[:,1], c=y, cmap='rainbow'); axs[0].set_title('original data')
axs[1].scatter(X[:,0], X[:,1], c=y_predict, cmap='rainbow'); axs[1].set_title('ANN prediction (accuracy = %.3f'%accuracy+')')
axs[0].grid(); axs[1].grid()
```

# 训练 digits 数据

```python
data = datasets.load_digits()
X, y = data.data, data.target
```

```python
model.fit(X, y)
y_predict = model.predict(X)
metrics.accuracy_score(y, y_predict)
```

```python
model_selection.cross_val_score(model, X, y)
```

##### 不同架构的表现

```python
scores = []
layer_sets = []
for hidden_layer_sizes in [
    (20,),
    (50,),
    (100,),
    (200,),
    (300,),
    (500,),
    (800,),
    (1000,),
    (100,10),
    (300,10),
    (500,10),
    (800,20),
    (100,20,10),
    (10,100),
    (10,20,100),
    ]:
    
    model = neural_network.MLPClassifier(hidden_layer_sizes)
    layer_sets.append(hidden_layer_sizes)
    score = model_selection.cross_val_score(model, X, y, cv=5)
    scores.append(score)
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(range(len(scores)), [xx.mean() for xx in scores], marker='p', c='r', lw=2, markersize=15);
ax.set_xticks(range(len(scores)))
ax.set_xticklabels([str(xx) for xx in layer_sets], rotation=30);
ax.set_xlabel('hidden layer size', fontsize=15)
ax.set_ylabel('cross validation score', fontsize=15)
ax.grid()
```
